### Handle Missing Data
Dropping Unnecessary Columns: Removing columns that aren’t useful.
Filling Missing Values:"Unknown" for text data, median for numeric data.

#### Filling missing values for categorical columns with "Unknown"

```python

name_basics.fillna("Unknown", inplace=True)
title_basics.fillna("Unknown", inplace=True)
title_crew.fillna("Unknown", inplace=True)
title_principals.fillna("Unknown", inplace=True)
title_akas.fillna("Unknown", inplace=True)
```

#### Filling missing values for numerical columns only in title_ratings

```python
numeric_cols = title_ratings.select_dtypes(include=['number']).columns
title_ratings[numeric_cols] = title_ratings[numeric_cols].fillna(title_ratings[numeric_cols].median())
```

#### Replacing '\\N' with NaN

```python
def replace_missing_values(df):
    return df.replace('\\N', np.nan)

title_basics = replace_missing_values(title_basics)
name_basics = replace_missing_values(name_basics)
title_principals = replace_missing_values(title_principals)
title_crew = replace_missing_values(title_crew)
title_ratings = replace_missing_values(title_ratings)
title_akas = replace_missing_values(title_akas)
```

#### Checking missing values

```python
def check_missing_values(df, name):
    print(f"\nMissing Values in {name}:")
    print(df.isnull().sum())

check_missing_values(title_basics, "Title Basics")
check_missing_values(name_basics, "Name Basics")
check_missing_values(title_principals, "Title Principals")
check_missing_values(title_crew, "Title Crew")
check_missing_values(title_ratings, "Title Ratings")
check_missing_values(title_akas, "Title Akas")
```

#### Dropping unnecessary columns

```python

title_basics = title_basics.drop(columns=['isAdult', 'endYear'])
name_basics = name_basics.drop(columns=['deathYear','birthYear','primaryProfession','knownForTitles'])
title_principals = title_principals.drop(columns=['ordering', 'job', 'characters'])
title_akas = title_akas.drop(columns=['language', 'region', 'types', 'attributes'])
title_crew = title_crew.drop(columns=['writers'])
```

```python
# Merge title_ratings with title_basics
merged_df = pd.merge(title_basics, title_ratings, on='tconst', how='left')

# Merge title_principals
merged_df = pd.merge(merged_df, title_principals, on='tconst', how='left')

# Merge title_crew
merged_df = pd.merge(merged_df, title_crew, on='tconst', how='left')

# Merge with name_basics using nconst
merged_df = pd.merge(merged_df, name_basics, left_on='nconst', right_on='nconst', how='left')

```

```python
# Merge title_akas using tconst and titleId as they represent the same identifier
merged_df = pd.merge(merged_df, title_akas, left_on='tconst', right_on='titleId', how='left')

# Drop redundant titleId column from title_akas
merged_df.drop(columns=['titleId'], inplace=True)
```

```python
# Merge title_basics with ratings
titles_with_ratings = pd.merge(title_basics, title_ratings, on='tconst', how='left')

# Convert startYear to numeric
titles_with_ratings['startYear'] = pd.to_numeric(titles_with_ratings['startYear'], errors='coerce')

# 1. Genre Popularity Over Time (Focus on Top 5 Genres)
genre_year = titles_with_ratings.explode('genres').groupby(['startYear', 'genres']).size().reset_index(name='count')
top5_genres = genre_year.groupby('genres')['count'].sum().nlargest(5).index
genre_year_top5 = genre_year[genre_year['genres'].isin(top5_genres)]

plt.figure(figsize=(12,6))
sns.lineplot(data=genre_year_top5, x='startYear', y='count', hue='genres', palette='tab10')
plt.title('Popularity of Top 5 Movie Genres Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genre', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Ensure 'directors' column is parsed correctly
title_crew['directors'] = title_crew['directors'].fillna('').astype(str).str.split(',')

# Merge and explode directors
top_directors = title_crew.explode('directors').merge(title_ratings, on='tconst', how='inner')

# Group by directors and calculate their average ratings
top_directors = top_directors.groupby('directors')['averageRating'].mean().reset_index()

# Sort and select top 10 directors
top_directors = top_directors.sort_values(by='averageRating', ascending=False).head(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(data=top_directors, x='averageRating', y='directors', palette='viridis')
plt.title('Top 10 Directors with Highest Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Director')
plt.show()
```

```python
# 6. Top 10 Most Voted Movies
top_voted = titles_with_ratings.nlargest(10, 'numVotes')[['primaryTitle', 'numVotes']]
plt.figure(figsize=(12,6))
sns.barplot(data=top_voted, x='numVotes', y='primaryTitle', palette='Blues_r')
plt.title('Top 10 Most Voted Movies')
plt.xlabel('Number of Votes')
plt.ylabel('Movie Title')
plt.show()
```

```python
# 8. Top Rated Movies by Decade
titles_with_ratings['decade'] = (titles_with_ratings['startYear'] // 10) * 10
decade_avg_rating = titles_with_ratings.groupby('decade')['averageRating'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=decade_avg_rating, x='decade', y='averageRating', marker='o')
plt.title('Top Rated Movies by Decade')
plt.xlabel('Decade')
plt.ylabel('Average Rating')
plt.show()
```
